```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import learning_curve
from sklearn.model_selection import train_test_split
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
data = pd.read_csv(url, sep=';')
data
```

```python
# Print column names
print(data.columns)
```

```python
#split the dataset into training and testing sets
x= data[["alcohol"]].values
y= data["quality"].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lin_regressor=LinearRegression()
lin_regressor.fit(x_train,y_train)
poly=PolynomialFeatures(degree=2)
x_poly_train=poly.fit_transform(x_train)
x_poly_test=poly.transform(x_test)
poly_regressor = LinearRegression()
poly_regressor.fit(x_poly_train,y_train)
```

```python
# predict the quality of wine for test data using both models
y_pred_lin = lin_regressor.predict(x_test)
y_pred_poly = lin_regressor.predict(x_poly_test[:,0].reshape(-1,1))

# print the performance metrics for both models
print('LinearRegression Metrics:')
mse_lin = mean_squared_error(y_test, y_pred_lin)
rmse_lin = np.sqrt(mse_lin)
r2_lin = r2_score(y_test, y_pred_lin)
print('MeanSquared Error:', mse_lin)
print('Root MeanSquared Error:', rmse_lin)
print('R-squared:', r2_lin)

print('Polynomial Regression Metrics:')
mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)
r2_poly = r2_score(y_test, y_pred_poly)
print('MeanSquared Error:', mse_poly)
print('Root MeanSquared Error:', rmse_poly)
print('R-squared:', r2_poly)
```

```python
#plot the learning curve for both models
train_sizes,train_scores_lin,test_scores_lin=learning_curve(lin_regressor,x,y,cv=5)
train_sizes,train_scores_poly,test_scores_poly=learning_curve(poly_regressor,x,y)
train_mean_lin=np.mean(train_scores_lin,axis=1)        
```
